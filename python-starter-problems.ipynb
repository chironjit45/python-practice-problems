## Problem-7: Write a function that takes 1 number as argument. The function should return “Fizz” if the number is divisible by 3, the function should return “Buzz” if the number is divisible by 5, the function should return “FizzBuzz” if the number is divisible by both 5 and 3, otherwise return “Not a Fizz-buzz number”

```python
# type casting
my_input = int(input("Enter a number: "))

if my_input % 3 == 0 and my_input % 5 ==0:
    print("FizzBuzz")
elif my_input % 3 ==0:
    print("Fizz")
elif my_input % 5 ==0:
    print("Buzz")
else:
    print("Not a FizzBuzz number")
```

# Problem-11: Find the most frequent character in the paragraph.
## rhyme = 'Twinkle, twinkle, little star. How I wonder what you are!

```python
from collections import Counter
rhyme = 'Twinkle, twinkle, little star. How I wonder what you are!'

clean_rhyme = ''.join(char.lower() for char in rhyme if char.isalpha())

char_count = Counter(clean_rhyme)
most_fre_char = char_count.most_common(1)[0]

print(f"The most frequent character is {most_fre_char[0]}, which repeact is {most_fre_char[1]} times.")
```

## Problem-12: Find the common items between the lists and make SUM of the new list items which are common between the lists. list1 = [3, 5, 7, 4, 8, 8], list2 = [4, 9, 8, 7, 1, 1, 13]

```python
list1 = [3, 5, 7, 4, 8, 8]

list2 = [4, 9, 8, 7, 1, 1, 13]

common_items = set(list1) & set(list2)
print("Common items:",common_items)
print("Sum = ",sum(common_items))
```

## Problem-13: Compare the two dictionaries and find the common items based on KEYs of the dictionaries. dict1 = {'age': 13, 'id': 12, 'address': 'Banani', 'course': 'Python'}, dict2 = {'address': 'Rupnagar', 'id': 25, 'course': 'MERN'}

```python
dict1 = {'age': 13, 'id': 12, 'address': 'Banani', 'course': 'Python'}

dict2 = {'address': 'Rupnagar', 'id': 25, 'course': 'MERN'}

common_key = dict1.keys() & dict2.keys()
common_key

common_items = {}
for key in common_key:
    common_items[key] = (dict1[key],dict2[key])

print("Common items and their values.")
print(common_items)
```

## Problem-14: Sort the list in DESCENDING order and find the subtraction of maximum and minimum numbers. list_1 = [4, 9, 8, 7, 5, 2, 13]

```python
list_1 = [4, 9, 8, 7, 5, 2, 13]
list_1.sort(reverse=True)
max_num= max(list_1)
min_num = min(list_1)

subtraction = max_num + min_num
print(f"Sorted list(Decending Order: {list_1} and the subtraction(max & min) is: {subtraction}")
```
